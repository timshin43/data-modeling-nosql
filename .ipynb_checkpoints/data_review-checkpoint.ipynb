```python
import boto3
import configparser
import pandas as pd
from time import time
```

```python
### AWS
config = configparser.ConfigParser()
config.read_file(open('dwh.cfg'))
KEY=config.get('AWS','KEY')
SECRET= config.get('AWS','SECRET')

DWH_DB= config.get("DWH","DWH_DB")
DWH_DB_USER= config.get("DWH","DWH_DB_USER")
DWH_DB_PASSWORD= config.get("DWH","DWH_DB_PASSWORD")
DWH_PORT = config.get("DWH","DWH_PORT")

DWH_ENDPOINT= config.get("CLUSTER","HOST")
DWH_ROLE_ARN = config.get("IAM_ROLE","ARN")
```

#### Review Logs Data

```python
# Review 
s3 = boto3.resource('s3',
                   region_name="us-west-2",
                   aws_access_key_id=KEY,
                   aws_secret_access_key=SECRET)

sampleDbBucket = s3.Bucket("udacity-dend")

for obj in sampleDbBucket.objects.filter(Prefix='log_data'):
    print(obj)
```

```python
s2TestFile = sampleDbBucket.objects.filter(bucket_name='udacity-dend',key='log_data/2018/11/2018-11-01-events.json')
```

```python
s3_client = boto3.client('s3',
                   region_name="us-west-2",
                   aws_access_key_id=KEY,
                   aws_secret_access_key=SECRET)
```

```python
response = s3_client.get_object(Bucket='udacity-dend', Key='log_data/2018/11/2018-11-01-events.json')
df = pd.read_json(response.get('Body'),lines=True)
df
```
